```python
import pandas as pd
import numpy as np
import json

bucket_src = "fcx-raw-data-temp"
path_to_file="FEGS/data"
filename = "goesr_plt_FEGS_20170321_Flash_v2.txt"
s3path=f"s3://{bucket_src}/{path_to_file}/{filename}"
print(s3path)
```

```python
subset = pd.read_csv(s3path, sep=",",index_col=False,usecols=
                  ['FlashID', 'peak'])

subset = pd.read_csv(s3path, sep=",", index_col='FlashID', usecols=['FlashID', 'peak'])
print(subset)
op = json.loads(subset.to_json(orient='split'))
```

```python
op['data'] = np.array(op['data']).flatten().tolist()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(op['index'], op['data'])
plt.show()
```
